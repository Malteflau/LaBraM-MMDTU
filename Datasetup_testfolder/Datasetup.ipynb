```python
import mne
import pandas as pd
import numpy as np

# Path to your file
file_path = "/Users/maltelau/Desktop/LaBraM-MMDTU/LaBraM-MMDTU/DTUDATA/FG_Data/PreprocessedEEGData/315A_FG_preprocessed-epo.fif"

# Method 1: Using to_data_frame()
epochs = mne.read_epochs(file_path)
df_method1 = epochs.to_data_frame()

# Method 2: Using get_data()
eeg_data = epochs.get_data()

# Let's compare the values
# Print first few values from both methods for the first epoch and channel
print("Method 1 (to_data_frame) first few values:")
print(df_method1.iloc[0:5, :])  # First 5 rows

print("\nMethod 2 (get_data) first few values:")
print(eeg_data[0, 0, 0:5])  # First epoch, first channel, first 5 time points

# Let's also check the shapes
print("\nShapes:")
print(f"Method 1 (to_data_frame) shape: {df_method1.shape}")
print(f"Method 2 (get_data) shape: {eeg_data.shape}")

# Print channel names
print("\nChannel names:")
print(epochs.ch_names)

# Create a comparison function
def compare_values(epoch_idx=0, channel_idx=0, time_points=5):
    """
    Compare values between the two methods for a specific epoch and channel
    """
    # Get values from method 1 (to_data_frame)
    channel_name = epochs.ch_names[channel_idx]
    epoch_data_df = df_method1[df_method1['epoch'] == epoch_idx][channel_name].iloc[:time_points]
    
    # Get values from method 2 (get_data)
    epoch_data_array = eeg_data[epoch_idx, channel_idx, :time_points]
    
    print(f"\nComparison for epoch {epoch_idx}, channel {channel_name}:")
    print("DataFrame method:", epoch_data_df.values)
    print("Array method:    ", epoch_data_array)
    
    # Check if values are equal (within numerical precision)
    is_equal = np.allclose(epoch_data_df.values, epoch_data_array)
    print("Values are equal:", is_equal)

# Compare the first epoch, first channel
compare_values(0, 0, 5)
```

```python
import mne
import pandas as pd
import numpy as np
import os
import pickle

# Define paths
folder_path = "/Users/maltelau/Desktop/LaBraM-MMDTU/LaBraM-MMDTU/DTUDATA/FG_Data/PreprocessedEEGData"
force_data_path = "/Users/maltelau/Desktop/LaBraM-MMDTU/LaBraM-MMDTU/DTUDATA/FG_Data/Force_df_v2.pkl"

# Load force data
with open(force_data_path, "rb") as f:
    force_df = pickle.load(f)

# Initialize lists to store data
all_data = []

# Get list of all .fif files
fif_files = [f for f in os.listdir(folder_path) if f.endswith('.fif')]
t=0
# Process each file
for fif_file in fif_files:
    t+=1
    if t==11:
        break
    # Extract participant ID from filename (assuming format like "301A_FG_preprocessed-epo.fif")
    participant_id = fif_file.split('_')[0]

    # Load epochs
    epochs = mne.read_epochs(os.path.join(folder_path, fif_file), preload=True)
    
    # Get data from epochs
    eeg_data = epochs.get_data()  # Shape: (n_epochs, n_channels, n_times)
    
    # For each epoch
    for epoch_idx in range(len(epochs)):
        # Get corresponding force data
        force_data = force_df[
            (force_df['Triad_id'] == int(participant_id[:3])) & 
            (force_df['Trial'] == epoch_idx + 1)
        ]
        
        if len(force_data) > 0:
            # Create row for each epoch
            row_data = {
                'participant_id': participant_id,
                'epoch_idx': epoch_idx,
                'condition': epochs.events[epoch_idx][2],  # Get condition from events
                'eeg_data': eeg_data[epoch_idx],  # Store full EEG data
                'force_level': force_data['Force_level'].iloc[0],
                'target_force': force_data['TargetForce'].iloc[0],
                'trial_time': force_data['Trial_time'].iloc[0],
                'force': force_data['Force'].iloc[0],
                'dforce': force_data['dForce'].iloc[0]
            }
            all_data.append(row_data)

# Create DataFrame
combined_df = pd.DataFrame(all_data)

# Example of filtering/accessing data
def get_condition_data(df, condition, participants=None):
    """
    Filter data by condition and optionally by participants
    """
    mask = df['condition'] == condition
    if participants:
        mask &= df['participant_id'].isin(participants)
    return df[mask]

# Example usage:
# Get all T23Pn trials
t23pn_data = get_condition_data(combined_df, 'T23Pn')

print(f"Combined DataFrame shape: {combined_df.shape}")
print("\nColumns:", combined_df.columns.tolist())
```
